```python
from typing import List

class Exercising:
    def __init__(self, name: str, qty: int) -> None:
       self._name = name
       self._qty = qty

    @property
    def name(self) -> str:
        return self._name

    @property
    def qty(self) -> int:
        return self._qty
  
    def __rpr__(self) -> str:
        return f"{self._name} {self._qty}"

def read_list(file_name: str) -> List[Exercising]:
    exercises: List[Exercising] = [] 
    with open(file_name, 'r') as f:
        exercises_qty = int(f.readline())
        for i in range(exercises_qty):
            name_ex, qty_ex = f.readline().split()
            exercises.append(Exercising(name_ex, int(qty_ex)))
    return exercises

exercises: List[Exercising] =  read_list('U2.txt')


for exercise in list(set([exercise.name for exercise in exercises])):
    done_ex_qty = 0
    for list_exercise in exercises:
        if list_exercise.name == exercise:
            done_ex_qty += list_exercise.qty
    print(exercise, done_ex_qty)

#with open('U2Rez.txt', 'w') as w:
 #   for key, value in
```

```python
from typing import List, Tuple

class Sheep:
    def __init__(self, name: str, dnr: str) -> None:
        self._name = name
        self._dnr = dnr
        self._coincidence_cof = 0 

    @property
    def dnr(self) -> str:
        return self._dnr

    @property
    def coincidence_cof(self) -> int:
        return self._coincidence_cof    

    @property
    def name(self) -> str:
        return self._name  

    @coincidence_cof.setter
    def coincidence_cof(self, cof: int) -> None:
        self._coincidence_cof = cof

    def coincidence(self, sheep) -> int:
        coincidence_cof = 0
        for i, letter in enumerate(self._dnr):
            if letter == sheep.dnr[i]:
                coincidence_cof += 1
        return coincidence_cof

    def __eq__(self, __o) -> bool:
        return self.coincidence_cof == __o.coincidence_cof and self.name == __o.name    

    def __lt__(self, __o) -> bool:
        if self.coincidence_cof == __o.coincidence_cof:
            return self.name > __o.name
        return self.coincidence_cof < __o.coincidence_cof        

def read_sheeps(file_name: str) -> Tuple[List[Sheep], Sheep]:
    all_sheep: List[Sheep] = []
    with open(file_name, 'r') as f:
        f.readline()
        examine_sheep_index = int(f.readline())
        for i, line in enumerate(f.readlines()):
            name, dnr = line.split()
            if i + 1 == examine_sheep_index:
                examine_sheep = Sheep(name, dnr)
            else:
                all_sheep.append(Sheep(name, dnr))
    return all_sheep, examine_sheep

all_sheep, examine_sheep = read_sheeps('UU2.txt')

for sheep in all_sheep:
    coincidence_cof = sheep.coincidence(examine_sheep)
    sheep.coincidence_cof = coincidence_cof

all_sheep.sort(reverse = True)

with open('UU2_rez.txt', 'w') as f:
    for sheep in all_sheep: #kreipiasi i __eq__ ir __lt__ jeigu sorted() naudoji
        f.write(f"{sheep.name}, {sheep.coincidence_cof} \n")    


```

```python
from typing import List, Dict

class Savivaldybe:
    def __init__(self, pavadinimas: str, mokyklos: int) -> None:
        self._pavadinimas = pavadinimas ## vidinis atributas
        self._mokyklos = mokyklos

    @property
    def pavadinimas(self) -> str:
        return self._pavadinimas

    @property
    def mokyklos(self) -> int:
        return self._mokyklos    


class Apskritis:
    def __init__(self, pavadinimas: str) -> None:
        self._pavadinimas = pavadinimas
        self._savivaldybes: List[Savivaldybe] = []

    @property
    def savivaldybiu_kiekis(self) -> int:
        return len(self._savivaldybes)

    @property
    def mokyklu_max_kiekis(self) -> int:
        return max([savivaldybe.mokyklos for savivaldybe in self._savivaldybes])

    def prideti_savivaldybe(self, savivaldybe: Savivaldybe) -> None:
        self._savivaldybes.append(savivaldybe)

    @property
    def pavadinimas(self) -> str:
        return self._pavadinimas    

def skaityti_duomenis(file_name: str) -> List[Apskritis]:
    apskriciu_sarasas: Dict[str, Apskritis] = {}
    with open(file_name, 'r') as f: # r skaitymo protokolas
        savivaldybiu_kiekis = int(f.readline())
        for i in range(savivaldybiu_kiekis):
            savivaldybes_pavadinimas, mokyklu_skaicius = f.readline().split()
            savivaldybe = Savivaldybe(savivaldybes_pavadinimas, int(mokyklu_skaicius))
            apskrities_pavadinimas = f.readline().replace('\n', '') # viena eilute skaitom, ir new line kad nebutu ir neskaitytu, nes sugadina duomenis
            if apskrities_pavadinimas in apskriciu_sarasas:
                apskriciu_sarasas[apskrities_pavadinimas].prideti_savivaldybe(savivaldybe)
            else:
                apskriciu_sarasas[apskrities_pavadinimas] = Apskritis(apskrities_pavadinimas)   
                apskriciu_sarasas[apskrities_pavadinimas].prideti_savivaldybe(savivaldybe) 
    return list(apskriciu_sarasas.values()) 

apskritys = skaityti_duomenis('U3.txt')
for apkritis in apskritys:
    print(apkritis.pavadinimas, apkritis.savivaldybiu_kiekis, apkritis.mokyklu_max_kiekis)           

open('U3.txt', 'r').close()
```

```python
from typing import Tuple, List
from datetime import datetime

class Slidininkas:
    def __init__(self, vardas: str, pavarde: str) -> None:
        self._vardas = vardas
        self._pavarde = pavarde
        self._starto_laikas: datetime = None
        self._finisho_laikas: datetime = None

    @property
    def vardas(self) -> str:
        return self._vardas

    @property
    def pavarde(self) -> str:
        return self._pavarde

    @property
    def starto_laikas(self) ->datetime:
        return self._starto_laikas

    @property
    def finisho_laikas(self) ->datetime:
        return self._finisho_laikas

    @starto_laikas.setter
    def starto_laikas(self, laikas: datetime) -> None:
        self._starto_laikas = laikas
    @property
    def vardas_pavarde(self) -> str:
        return f"{self._vardas} {self._pavarde}"

    @starto_laikas.setter
    def starto_laikas(self, laikas: datetime) -> None:
        self._starto_laikas = laikas    

    @finisho_laikas.setter
    def finisho_laikas(self, laikas: datetime) -> None:
        self._finisho_laikas = laikas 

     #funkcija esanti klaseje, kuri neiima vidiniu atributu
    @staticmethod
    def konvertuoti_laika(laikas: str) -> datetime:
        #print(f"*{laikas}*")
        return datetime.strptime(laikas.replace(" ",":"), "%H:%M:%S")

    def slidininko_trukme(self) ->int:
        return (self._finisho_laikas - self._starto_laikas).total_seconds()

    def __eq__(self, __o: object) -> bool:
        return self.slidininko_trukme() == __o.slidininko_trukme() and self.vardas_pavarde == __o.vardas_pavarde

    def __lt__(self, __o: object) -> bool: #kuris maziau
        if self.slidininko_trukme() == __o.slidininko_trukme():
            return self.vardas_pavarde < __o.vardas_pavarde
        return self.slidininko_trukme() < __o.slidininko_trukme()

    @staticmethod
    def formatuoti_laika(laikas: int) -> str:
        return f"{laikas // 60} {laikas % 60}"

def istraukti_duomenis(eilute: str) -> Tuple[str, str, str]:
    vardas_pavarde = eilute[:20]
    laikas = eilute[20:]
    pavarde = vardas_pavarde.split()[-1]
    vardas =  " ".join(vardas_pavarde.split()[:-1])
    #print(vardas, pavarde, laikas)
    return vardas, pavarde, laikas

def skaityti_faila(pavadinimas) -> List[Slidininkas]:
    slidininkai: List[Slidininkas] = []
    with open(pavadinimas, "r") as f:
        startavusiu_slidininku_sk = int(f.readline())
        for _ in range(startavusiu_slidininku_sk):
            eilute = f.readline().rstrip()
            vardas, pavarde, laikas = istraukti_duomenis(eilute)
            slidininkas = Slidininkas(vardas, pavarde)
            slidininkas.starto_laikas = Slidininkas.konvertuoti_laika(laikas)
            slidininkai.append(slidininkas)
        finisavusiu_slininku_sk = int(f.readline())
        for _ in range(finisavusiu_slininku_sk):
            eilute = f.readline().rstrip()  
            vardas, pavarde, laikas = istraukti_duomenis(eilute)  
            for f_slidininkas in slidininkai:
                if f_slidininkas.vardas == vardas and f_slidininkas.pavarde == pavarde:
                    f_slidininkas.finisho_laikas = Slidininkas.konvertuoti_laika(laikas)
                    #print(slidininkas.finisho_laikas)
                    break 
        return slidininkai

slidininkai = skaityti_faila("rez.txt")

finishave_slidininkai = [slidininkas for slidininkas in slidininkai if slidininkas.finisho_laikas]

for slidininkas in sorted(finishave_slidininkai):
    print(slidininkas.vardas, slidininkas.pavarde, Slidininkas.formatuoti_laika(slidininkas.slidininko_trukme()))  
```
